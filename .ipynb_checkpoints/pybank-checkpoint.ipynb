```python
import numpy as np
import pandas as pd
```

```python


df = pd.read_csv('./Resources/budget_data.csv')


# The total number of months included in the dataset.

total_months = len(df)


# The net total amount of Profit/Losses over the entire period.

total_profit_losses = df['Profit/Losses'].sum()

# The average of the changes in Profit/Losses over the entire period.

difference = []
previous_number = 0
current_number = 0
for i in df.values:
    if previous_number == 0:
        previous_number = i[1]
    elif previous_number != 0 and current_number == 0:
        current_number = i[1]
    if current_number != 0 and previous_number != 0:
        difference.append([i[0],current_number-previous_number])
        previous_number = current_number
    current_number = i[1]
        

# The greatest increase in profits (date and amount) over the entire period.

pd.DataFrame(difference).nlargest(1,1)

# The greatest decrease in losses (date and amount) over the entire period

pd.DataFrame(difference).nsmallest(1,1)


print("Financial Analysis 1")
print("----------------------------")
print("Total months:",total_months)
print('Total:', total_profit_losses)
print (pd.DataFrame(difference).nlargest(1,1))
print("Greatest Decrease in Profits")
print(pd.DataFrame(difference).nsmallest(1,1))

```
